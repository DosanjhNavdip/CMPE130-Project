```python
import csv
import sys
from datetime import datetime


def helper(arr, x, l, r):
    found = True
    for i in range(l, r):
        check = True;
        for j in range (len(x)):
            if x[j].capitalize() != arr[i][1][j].capitalize():
                check = False;
                break
        if check:
            if found:
                print("Book", "\t\t\t\t" ," Author", "\t\t\t\t", "ISBn13")
            print(arr[i][1], "\t", arr[i][2], "\t", arr[i][5])
            found = False
    if found:
        print(x, " not found.")
def binarySearch(arr, x):
    l = 0
    r = len(arr)
    check = 0
    check2 = 0
    while (l <= r):
        m = l + ((r - l) // 2)

        # Check if x is present at mid
        if  (" " + x + " ") in (" "+ arr[m][1] + " "):
            return m - 1

        if x[0] == arr[m][1][0]:
            helper(arr, x, l, r)
            return
        # If x greater, ignore left half
        if x[0] > arr[m][1][0]:
            l = m + 1

        # If x is smaller, ignore right half
        else:
            r = m - 1

    return -1

# Merge sort in Python
def mergeSort(nlist):
  if len(nlist) > 1:
    mid = len(nlist) // 2
    lefthalf = nlist[:mid]
    righthalf = nlist[mid:]

    mergeSort(lefthalf)
    mergeSort(righthalf)
    i = j = k = 0
    while i < len(lefthalf) and j < len(righthalf):
      if str(lefthalf[i]).split(",")[1] < str(righthalf[j]).split(",")[1] :
        nlist[k] = lefthalf[i]
        i = i + 1
      else:
        nlist[k] = righthalf[j]
        j = j + 1
      k = k + 1

    while i < len(lefthalf):
      nlist[k] = lefthalf[i]
      i = i + 1
      k = k + 1

    while j < len(righthalf):
      nlist[k] = righthalf[j]
      j = j + 1
      k = k + 1
  return nlist

def splitArray(arr):
    my_arr = []

    for i in arr:
        temp = str(i).split(',')
        temp[1] = temp[1].strip()
        temp[1] = temp[1].capitalize()
        temp.pop(-1)
        my_arr.append(temp)

    return my_arr

def arrPrint(arr):
    for i in arr:
        print (i)

with open("booksdatabasenew2.csv", 'r', encoding="utf-8") as file:
    content = file.readlines()
header = content[:1]
rows = content[1:]

header = splitArray(header)
rows = splitArray(rows)

arrPrint(header)

t1 = datetime.now()
row = mergeSort(rows)

t2 = datetime.now()
T = t2 - t1
print ("Time Taken for Sorting: ", T)
print("Enter Title to search: ")
x = input()

t1 = datetime.now()
binarySearch(row, x)
t2 = datetime.now()
T = t2 - t1
print ("Time Taken for Searching: ", T)

```
